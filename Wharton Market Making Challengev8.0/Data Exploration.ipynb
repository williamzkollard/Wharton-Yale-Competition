```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Define the correct order of credit ratings
credit_rating_order = ["CCC", "B", "BB", "BBB", "A", "AA", "AAA"]

# Convert "credit_rating" to categorical with the correct order
historical_bonds["credit_rating"] = pd.Categorical(historical_bonds["credit_rating"], categories=credit_rating_order, ordered=True)

# Sort the dataframe by region first, then by credit rating
historical_bonds = historical_bonds.sort_values(by=["region", "credit_rating"])

# Create a new column combining region and credit rating
historical_bonds["credit_region"] = historical_bonds["region"] + " - " + historical_bonds["credit_rating"].astype(str)

# Set figure size
plt.figure(figsize=(12, 24))

# First plot: Cash Flow 1 distribution by credit rating & region
plt.subplot(3, 1, 1)
sns.boxplot(x="credit_region", y="cash_flow_1", data=historical_bonds, order=historical_bonds["credit_region"].unique())
plt.xticks(rotation=45, ha="right")  # Rotate x labels for better visibility
plt.title("Distribution of Cash Flow 1 by Credit Rating & Region")
plt.xlabel("Credit Rating & Region")
plt.ylabel("Cash Flow 1")
plt.grid(True)

# Second plot: Cash Flow 2 distribution by credit rating & region
plt.subplot(3, 1, 2)
sns.boxplot(x="credit_region", y="cash_flow_2", data=historical_bonds, order=historical_bonds["credit_region"].unique())
plt.xticks(rotation=45, ha="right")
plt.title("Distribution of Cash Flow 2 by Credit Rating & Region")
plt.xlabel("Credit Rating & Region")
plt.ylabel("Cash Flow 2")
plt.grid(True)

# Third plot: Cash Flow 3 distribution by credit rating & region
plt.subplot(3, 1, 3)
sns.boxplot(x="credit_region", y="cash_flow_3", data=historical_bonds, order=historical_bonds["credit_region"].unique())
plt.xticks(rotation=45, ha="right")
plt.title("Distribution of Cash Flow 3 by Credit Rating & Region")
plt.xlabel("Credit Rating & Region")
plt.ylabel("Cash Flow 3")
plt.grid(True)

# Adjust layout to avoid overlap
plt.tight_layout()

# Show plots
plt.show()
```
