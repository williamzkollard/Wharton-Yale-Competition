```python

# One-hot encode the categorical variables
data = pd.get_dummies(historical_bonds, columns=['credit_rating', 'region'], drop_first=True)

# Separate features and target variable
X = data.drop(['cash_flow_1','cash_flow_2', 'cash_flow_3'], axis=1)
y = data['cash_flow_1']

poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
X_interact = poly.fit_transform(X)


esg_squared = (X['esg_rating'] ** 2).values.reshape(-1, 1)

interaction_with_esg_sq = []
for col in X.columns:
    if col != 'esg_rating':
        interaction = (X['esg_rating'] ** 2) * X[col]
        interaction_with_esg_sq.append(interaction.values.reshape(-1, 1))
        
# Combine the manually created interaction terms into a single array (if any exist)
if interaction_with_esg_sq:
    interactions_esg_sq = np.hstack(interaction_with_esg_sq)
    # Combine all features: original interactions, esg_rating², and the new interactions with esg_rating²
    X_combined = np.hstack([X_interact, esg_squared, interactions_esg_sq])
else:
    # If there are no other features to interact with, just combine X_interact and esg_squared
    X_combined = np.hstack([X_interact, esg_squared])



    
model = LinearRegression()
model.fit(X_combined, y)

# Predict and evaluate the model
y_pred = model.predict(X_combined)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("Mean Squared Error:", mse)
print("R² Score:", r2)
np.sqrt(mse)


```

```python
import joblib

# Save the model and any transformers needed for preprocessing
joblib.dump(model, 'cash_flow_1_model.pkl')
joblib.dump(poly, 'cash_flow_1_poly.pkl')
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
X = historical_bonds[["debt_to_equity", "energy_capacity", "esg_rating"]]
y = historical_bonds["true_bond_value"]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# --- 3. Build the feed-forward neural network ---
model = Sequential([
    Dense(64, input_dim=3, activation='relu'),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')  # Sigmoid activation for binary classification
])

# Compile the model with Adam optimizer and binary crossentropy loss
model.compile(optimizer=Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# --- 4. Train the model ---
model.fit(X_train, y_train, epochs=500, batch_size=32, validation_split=0.2)

# --- 5. Evaluate the model ---
loss, accuracy = model.evaluate(X_test, y_test)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
```
