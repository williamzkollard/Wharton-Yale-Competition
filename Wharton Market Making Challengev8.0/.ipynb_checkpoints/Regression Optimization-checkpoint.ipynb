```python
training_data = historical_bonds[["energy_capacity","credit_region", "debt_to_equity", "esg_rating", "cash_flow_1", "cash_flow_2", "cash_flow_3"]]
training_data
```

```python
grouped = training_data.groupby('credit_region')
regression_results = []

for region, group in grouped:
    X = group[['energy_capacity']].values  
    y = group['cash_flow_1'].values 
    
    model = LinearRegression()
    model.fit(X, y)
    
    regression_results.append({
        'credit_region': region,
        'gradient': model.coef_[0],
        'intercept': model.intercept_,
        'cash_flow_model': 0
    })

cash_flow_1_models = pd.DataFrame(regression_results)
print(cash_flow_1_models)
```

```python

#Training regressions with independent variable as energy_capacity 

grouped = training_data.groupby('credit_region')
regression_results = []

for region, group in grouped:
    X = group[['energy_capacity']].values  
    y = group['cash_flow_1'].values 
    
    model = LinearRegression()
    model.fit(X, y)
    
    regression_results.append({
        'credit_region': region,
        'gradient': model.coef_[0],
        'intercept': model.intercept_,
        'cash_flow_model': 0
    })

cash_flow_1_models = pd.DataFrame(regression_results)





#Plotting results and computing MSE

merged_df = historical_bonds.merge(cash_flow_1_models, on='credit_region', how='left')
merged_df['cash_flow_1_prediction'] = (merged_df['energy_capacity'] * merged_df['gradient']) + merged_df['intercept']

num_observations = 1000
actual = merged_df['cash_flow_1']
predicted = merged_df['cash_flow_1_prediction']  


mse = np.mean((actual - predicted) ** 2)
x = np.arange(1, num_observations + 1)
plt.figure(figsize=(30, 5))


for i in range(num_observations):
    plt.plot([x[i], x[i]], [actual[i], predicted[i]], color='gray', linestyle='-', alpha=0.5)


plt.scatter(x, actual, color='blue', marker='o', label="Actual", alpha=0.8)
plt.scatter(x, predicted, color='orange', marker='s', label="Predicted", alpha=0.8)

plt.xlabel("Observation")
plt.ylabel("Value")

plt.legend()
plt.title(f"Actual vs. Predicted Values\nIndependent Variable : energy_capacity\nMean Squared Error (MSE): {mse:.2f}")
plt.show()
```

```python
grouped = training_data.groupby('credit_region')
regression_results = []

for region, group in grouped:
    X = group[['energy_capacity', 'debt_to_equity']].values  # Now includes both independent variables
    y = group['cash_flow_1'].values 
    
    model = LinearRegression()
    model.fit(X, y)
    
    regression_results.append({
        'credit_region': region,
        'energy_capacity_coefficient': model.coef_[0],
        'debt_to_equity_coefficient': model.coef_[1],
        'intercept': model.intercept_,
        'cash_flow_model': 0  # Placeholder
    })

cash_flow_2_models = pd.DataFrame(regression_results)




#Plotting results and computing MSE

merged_df_2 = historical_bonds.merge(cash_flow_2_models, on='credit_region', how='left')
merged_df_2['cash_flow_1_prediction'] = merged_df_2['energy_capacity'] * merged_df_2['energy_capacity_coefficient'] + merged_df_2['debt_to_equity'] * merged_df_2['debt_to_equity_coefficient']+ merged_df_2['intercept']

num_observations = 1000
actual = merged_df_2['cash_flow_1']
predicted = merged_df_2['cash_flow_1_prediction']  


mse = np.mean((actual - predicted) ** 2)
x = np.arange(1, num_observations + 1)
plt.figure(figsize=(30, 5))


for i in range(num_observations):
    plt.plot([x[i], x[i]], [actual[i], predicted[i]], color='gray', linestyle='-', alpha=0.5)


plt.scatter(x, actual, color='blue', marker='o', label="Actual", alpha=0.8)
plt.scatter(x, predicted, color='orange', marker='s', label="Predicted", alpha=0.8)

plt.xlabel("Observation")
plt.ylabel("Value")

plt.legend()
plt.title(f"Actual vs. Predicted Values\nIndependent Variable : energy_capacity & debt_to_equity\nMean Squared Error (MSE): {mse:.2f}")
plt.show()

```

```python
grouped = training_data.groupby('credit_region')
regression_results = []

for region, group in grouped:
    X = group[['energy_capacity', 'debt_to_equity', 'esg_rating']].values  # Now includes both independent variables
    y = group['cash_flow_1'].values 
    
    model = LinearRegression()
    model.fit(X, y)
    
    regression_results.append({
        'credit_region': region,
        'energy_capacity_coefficient': model.coef_[0],
        'debt_to_equity_coefficient': model.coef_[1],
        'esg_rating_coefficient': model.coef_[2],
        'intercept': model.intercept_,
        'cash_flow_model': 0  # Placeholder
    })

cash_flow_3_models = pd.DataFrame(regression_results)




#Plotting results and computing MSE

merged_df_3 = historical_bonds.merge(cash_flow_3_models, on='credit_region', how='left')
merged_df_3['cash_flow_1_prediction'] = merged_df_3['energy_capacity'] * merged_df_3['energy_capacity_coefficient'] + merged_df_3['debt_to_equity'] * merged_df_3['debt_to_equity_coefficient']+ merged_df_3['esg_rating'] * merged_df_3['esg_rating_coefficient']+ merged_df_3['intercept']

num_observations = 1000
actual = merged_df_3['cash_flow_1']
predicted = merged_df_3['cash_flow_1_prediction']  


mse = np.mean((actual - predicted) ** 2)
x = np.arange(1, num_observations + 1)
plt.figure(figsize=(30, 5))


for i in range(num_observations):
    plt.plot([x[i], x[i]], [actual[i], predicted[i]], color='gray', linestyle='-', alpha=0.5)


plt.scatter(x, actual, color='blue', marker='o', label="Actual", alpha=0.8)
plt.scatter(x, predicted, color='orange', marker='s', label="Predicted", alpha=0.8)

plt.xlabel("Observation")
plt.ylabel("Value")

plt.legend()
plt.title(f"Actual vs. Predicted Values\nIndependent Variable : energy_capacity & debt_to_equity & esg_capacity \nMean Squared Error (MSE): {mse:.2f}")
plt.show()
```
